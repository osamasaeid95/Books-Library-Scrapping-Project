```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://books.toscrape.com/catalogue/category/books/mystery_3/index.html")
response
```

```python
soup = BeautifulSoup(response.text)
```

```python
articles = soup.find_all('article', class_="product_pod")
articles
```

```python
# Check one article
articles[0]
```

```python
art = articles[0]

# Part 1: Title
h3 = art.find('h3')  # a = art.find('h3').find('a')
a = h3.find('a')

title = a.attrs['title']

# Part 2: Price
price = art.find('p', class_="price_color").text
price = float(price[2:])

# Part 3: Rating
rating = art.find('p', class_="star-rating")
rating = rating.attrs['class'][-1]

title, price, rating
```

```python
def convert(num):
    
    values = {
        'One':1,
        'Two':2,
        'Three':3,
        'Four':4,
        'Five':5
    }

    return values[num]
```

```python
for art in articles:
    # Part 1: Title
    h3 = art.find('h3')  # a = art.find('h3').find('a')
    a = h3.find('a')

    title = a.attrs['title']

    # Part 2: Price
    price = art.find('p', class_="price_color").text
    price = float(price[2:])

    # Part 3: Rating
    rating = art.find('p', class_="star-rating")
    rating = rating.attrs['class'][-1]

    num_rating = convert(rating)

    print(f"{title}, {price}, {num_rating}")
```
